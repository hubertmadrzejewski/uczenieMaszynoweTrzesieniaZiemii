```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
earthquake = pd.read_csv('database.csv')
earthquake.head(-40)
earthquake.info()
earthquake['Magnitude'].plot.kde()
correlations = earthquake.corr()
fig, ax = plt.subplots(figsize=(12, 12))

colormap = sns.color_palette("BrBG", 12)

sns.heatmap(correlations, 
    cmap=colormap, 
    annot=True, 
    fmt=".2f")
ax.set_yticklabels(earthquake.columns);

plt.show()
```
